```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df.head()
```

```python
df = df.set_index('Countries')
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis')
```
