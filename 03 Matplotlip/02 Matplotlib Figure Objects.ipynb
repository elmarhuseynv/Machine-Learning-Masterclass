```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y)

plt.show()
```

```python
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(a, b)

plt.show()
```

```python
fig = plt.figure()


axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel('B')
axes1.set_title("Power of 4")

axes2 = fig.add_axes([0.2, 0.5,0.25,0.25])
axes2.plot(x,y)

plt.show()
```
